#**Visualization**

```python
plt.figure(figsize=(10,6))
plt.hist(cereal_data['calories'], bins=20, color='orange', edgecolor='black')
plt.title('Distribution of Calories in Cereals')
plt.xlabel('Calories per Serving')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(cereal_data['sugars'], cereal_data['fiber'], color='red', alpha=0.5)
plt.title('Sugar vs Fiber in Cereals')
plt.xlabel('Sugar Content')
plt.ylabel('Fiber Content')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(cereal_data['calories'], cereal_data['rating'], color='purple', alpha=0.5)
plt.title('Calories vs Cereal Rating')
plt.xlabel('Calories per Serving')
plt.ylabel('Cereal Rating')
plt.show()
```

#**Steps to Analysze Fruity Pebbles**

```python
fruity_pebbles = cereal_data[cereal_data['name'] == 'Fruity Pebbles']
fruity_pebbles
```

```python
# Calculate average values for all cereals
average_values = cereal_data[['calories', 'sugars', 'protein', 'fat', 'fiber']].mean()

# Combine Fruity Pebbles and average values into a DataFrame for easy comparison
comparison_df = pd.DataFrame({
    'Fruity Pebbles': fruity_pebbles[['calories', 'sugars', 'protein', 'fat', 'fiber']].values.flatten(),
    'Average': average_values.values
}, index=['Calories', 'Sugars', 'Protein', 'Fat', 'Fiber'])

# Plot comparison
comparison_df.plot(kind='bar', figsize=(10, 6), color=['orange', 'gray'])
plt.title('Nutritional Profile: Fruity Pebbles vs. Average Cereal')
plt.ylabel('Content per Serving')
plt.xticks(rotation=0)
plt.legend(loc='upper right')
plt.show()
```
